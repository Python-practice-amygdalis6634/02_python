```python
# add() 함수 호출:  3
# type(add):  <class 'function'>
# <function add at 0x0000021066458680>
# minus() 함수 호출:  2
# 12321
# hello hello

def add(a, b):
    return a + b

print("add() 함수 호출: ", add(1, 2))
print(type(add))
print(add)

def minus(a, b):
    return a - b

print("minus() 함수 호출: ",minus(3, 1))
print(type(minus))
print(minus)

def multiply(x, y):
    return x * y

print(multiply(111,111))

```

```python
def bar():
    print("bar()")
    return 0

func = bar

def runner(x):
    print(x())

runner(func)

```

```python
def add_all(*args):
    print("args에 담긴 모양: ", args)

    total = 0
    for num in args:
        total += num
    return total

print(add_all(10, 20, 30, 40, 50))
```

```python
def test5(**kwargs):
    print(kwargs, type(kwargs))

test5()
test5(menu = '돈까스', people = 3, price = 10_000)
```
