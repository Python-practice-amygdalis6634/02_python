```python
def kilhwan() -> list[int]:
    numbers = [n for n in range(10)]
    return numbers

print(kilhwan())
```

```python
def seyoung() ->list[tuple[int, int]]:
    you : list[tuple[int, int]] = []
    for plus in range(1,11):
        for minus in range(51, 61):
            you.append((plus, minus))
    return you

print(seyoung())

```

```python
def seyoung() -> dict[int, int]:
    return {k : k ** 2 for k in range(1, 6)}

print(seyoung())
```

```python
# list -> dict:  {'홍길동': 20, '신사임당': 40, '이순신': 30}

def make_age_map() -> dict[str, int]:
    name = ['홍길동', '신사임당', '이순신']
    age = [20, 40, 30]

    return dict(zip(name, age))

print('list -> dict: ', make_age_map())
```
