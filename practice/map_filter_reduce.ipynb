```python
# [{'name': '논개', 'score': 99}, {'name': '강감찬', 'score': 91}
# #[{'name': '논개', 'score': 99}, {'name': '강감찬', 'score': 91}]

students = [
    {'name': '윤세영', 'score': 90},
    {'name': '김길환', 'score': 85},
    {'name': '손흥민', 'score': 75},
    {'name': '김연아', 'score': 95},
]

students1 = list(filter( lambda x : x['score'] >=90, students))
students2 = [x for x in students if x['score'] >= 90 ]

# 부가설명
# comprehension에서 해석 단위를 쪼개어서 보자.

# 1번 묶음은 출력을 위한 맨 앞의 x
# 2번 묶음은 for ~ in 구문
# 3번 묶음은 if 절

# 즉, 우리가 생각해볼 수 있는것은 우리가 배운 단위로 묶어서 생각하면 되지 않을까?

print(students1)
print(students2)


# comprehension 다시 이해하기

# A for x in 변수
# -> 변수에서 원소들을 하나씩 꺼내서 x에 대입하고 그 원소들을 어떤 형태(A)로 출력할건지

fruits = ['귤', '오렌지', '유자', '레몬']


A = [x for x in fruits]

print("너무 좋아", A)

# reduce 함수 = reduce(함수, iterable, 초기값)

num : list[int] = [11, 22, 33, 44, 55, 66, 77, 88, 99]

total = reduce(lambda sum, curr : curr + sum, num, 0)

print(total)
```
