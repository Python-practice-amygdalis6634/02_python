```python
i = 0
while True:
    speed = input("""
        ----------------
        무엇을 실행할까요?
        번호를 눌러주세요.
        0. 시동 걸기
        1. 엑셀
        2. 브레이크
        3. 시동 끄기
        ----------------
    """)

    match speed:
        case '0':
            if i == 0:
                print("시동이 켜졌습니다.")
            else:
                print("주행중입니다.")

        case '1':
            if i <= 190:
                i += 10
                print("현재 속도: ", i, "km/h 입니다.")
            else:
                print("최대 속도입니다.")

        case '2':
            if i >= 10:
                i -= 10
                print("현재 속도: ", i, "km/h 입니다.")
            else:
                print("자동차가 멈춰있습니다.")

        case '3':
            if i == 0:
                print("시동이 꺼졌습니다.")
                break
            else:
                print("시동을 끌 수 없습니다.")


```
