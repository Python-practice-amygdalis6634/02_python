## map
iterable객체의 각요소에 대해 지정한 함수를 적용하고, 그 결과를 map객체로 반환
```
map(function, iterable) -> map
```




````python
nums: list = [1, 2, 3, 4, 5]

result = map(lambda x: x ** 2, nums)   # iterable -> list, set, str...
print(result, type(result))   # map 주소, map 타입
print("list(result): ", list(result))

# map() : 기존 list의 요소를 하나씩 꺼내 함수를 호출,
#         함수 결과들을 모아 새로운 list 생성
````

````python
# 문자열 리스트 -> 문자열 길이 리스트
````

````python
products: list[str] = ['아이스 아메리카노', '수박주스', '말차라떼', '식혜', '딸기 스무디']

products_len: list[int] = map(lambda x: len(x), products)
print(list(products_len))
````

````python
# 여러 순회 기능 객체(list) 통합
mid_term = [90, 55, 70]
final_term = [100, 40, 75]

avg_scores = list(map(lambda x, y: (x + y)/ 2, mid_term, final_term)  ) # map(콜백함수, iterable)
print(avg_scores)
````

## filter

주어진 iterable객체에 대해서 참/거짓 평가후 참인 요소들만 결과값으로 반환

```
filter(function, iterable) -> filter
```

````python
# n % 2 == 0
nums = list(range(1, 11))

evens = filter(lambda n: n % 2 == 0, nums)
odds = filter(lambda n: n % 2 == 1, nums)
threes = filter(lambda n: n % 3 == 0, nums)

print(evens, type(evens))
print("list(evens): ",list(evens))
print("list(odds): ",list(odds))
print("list(threes): ",list(threes))
````

````python
# 이름 목록에서 성이 홍씨인 사람만 찾아내기
names: list[str] = ['홍길동', '고길동', '홍진경', '최홍만', '홍명보']

# str.startswith(str) -> bool
# - 문자열의 시작이 str2와 같다면 True
hong_family = list(filter(lambda name: name.startswith('홍'), names))
print(hong_family)
````

````python
# dict 검색(dict 필터링)
# 학생 목록에서 점수가 90점 이상인 학생만 필터링
students = [
    { 'name' : '이순신', 'score' : 80},
    { 'name' : '윤봉길', 'score' : 75},
    { 'name' : '논개', 'score' : 99},
    { 'name' : '강감찬', 'score' : 91},
]

# filter() 함수
students_1 = list(filter(lambda s: s['score'] >= 90, students))
print(students_1)

# list comprehension
students_2 = [s for s in students if s['score'] >= 90]   # 리스트를 만들기 위해 마지막 대괄호
print(students_2)

````

## reduce
iterable객체의 요소에 대한 연산결과를 단일값을 반환.
합계, 최대/최소값 등의 연산에 적합

```
reduce(function, iterable, initial_value) -> Any
```
함수는 3개의 인자를 받는다.
1. 누적 연산값
2. 현재 요소
3. 초기값 (선택)


````python
# 외부 functools 모듈에서 reduce만 가져온다  -> 파이썬 API
from functools import reduce

# 정수형 list 내 요소들의 합계
nums: list[int] = [11, 22, 33, 44, 55, 66, 77, 88, 99]

# 총합 저장
# total = reduce(콜백함수(누적용변수, 현재값), 순회객체, 누적용변수 초기값)   누적용변수 초기값에는 아무런 영향이 없는 값을 넣어야함
# 여기서는 0인데 만약에 곱하기를 할 때는 1이 되겠지
# -> 콜백 함수의 return 값이 자동으로 누적용 변수에 대입
total = reduce(lambda sum, curr: sum + curr, nums, 0)
print("total: ", total)
````

````python
# 최대, 최소 값 찾기
nums: list[int] = [32, 123, -43, 3, 10, -34234, 33333, -556654, 1111]

# max_num: int = reduce(lambda max, curr: T if 조건 else F, nums, 0)
max_num: int = reduce(lambda max, curr: max if max > curr else curr, nums, 0)
print("최대값: ", max_num)

min_num: int = reduce(lambda min, curr: min if min < curr else curr, nums, 0)
print("최소값: ", min_num)

# 절대 값이 가장 큰 수 + 일반 함수
def abs_func(max, curr):
    print(f'max: {max}, curr: {curr}')
    return max if abs(max) > abs(curr) else curr

abs_max = reduce(abs_func, nums, 0)
print("abs_max: ", abs_max)
````

````python
# 최고 점수를 지닌 학생 출력
students = [
    { 'name' : '이순신', 'score' : 80},
    { 'name' : '윤봉길', 'score' : 75},
    { 'name' : '논개', 'score' : 99},
    { 'name' : '강감찬', 'score' : 91},
]

# 초기 값 생략 -> 첫 번째 값 대입

func = lambda max, curr: max if max['score'] > curr['score'] else curr
best_student = reduce(func, students)

print("결과: ", best_student)
````
